# Prediction of marks using supervised ML

For the first step, we import important libraries such as Panda for data analysis, NumPy for mathematical functions and mathplotlib to construct graphs.

```python
import pandas as pd  
import numpy as np    
import matplotlib.pyplot as plt  
%matplotlib inline
```

For the second step, we load the data through the URL provided with the help of Panda and the data is printed to make sure it has loaded successfully.

```python
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
data = pd.read_csv(url)
print(data)
```

For the next step we plot the data with Hours as the x-axis and Scores as the y-axis along with the respective labels. I have selected style fivethirtyeight for the graph.

```python
plt.style.use('fivethirtyeight')
data.plot(x="Hours", y="Scores", style="o")
plt.title("Hours Vs Percentage")
plt.ylabel("Percentage obtained")
plt.xlabel("Hours studied")
plt.show()
```

Now we divide the data into attributes and labels.

```python
X = data.iloc[:, :-1].values
Y = data.iloc[:, 1].values
```

Here we use Sci-kit learn's method of train_test_split to split the data into training and test sets. The training set will be used to train the data and the test set will be used to predict the data.

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test=train_test_split(X, Y,test_size=0.2, random_state=0)
```

In this step we train the algorithm.

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, Y_train)
```

The regression line is plotted in this step.

```python
line=reg.coef_*X+reg.intercept_
plt.scatter(X,Y)
plt.plot(X, line);
plt.show()
```

We predicted the scores in this step.

```python
print(X_test)
Y_prediction = reg.predict(X_test)
```

We compare the actual marks with the predicted marks over here to test the fit of our model.

```python
df = pd.DataFrame({"Actual Marks": Y_test, "Predicted Marks": Y_prediction})
df
```

Now to conclude this assignment and to test whether our model is successfull or not, we find out the predicted score for 9.25 hours studied. The answer we get is ~93.7%

```python
hours_studied = [[9.25]]
pred=reg.predict(hours_studied)
print("Number of hours={}".format(hours_studied))
print("Predicted score={}".format(pred[0]))
```
